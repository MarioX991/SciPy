```python
# Instead of using the pandas function, maybe we want to use some of ours oir
# maybe we want to include some of numpy funcitons as are standart diviation, varance..
# For this we are using the aggreegation 
```

```python
import numpy as np
```

```python
df.groupby(["year"])["lifeExp"].agg(np.std)
```

```python
df.groupby(["year"])["pop"].agg(np.mean)
```

```python
# How does average life expectancy and the GDP per capital changes
#over each year and continent
```

```python
df.groupby(["year","continent"])[["lifeExp","gdpPercap"]].agg(np.mean)
```

```python
df.groupby(["year","continent"])[["lifeExp","gdpPercap"]].agg([np.mean,np.std])
```

```python
# to pretty - multiple index display  
groupby = df.groupby(["year","continent"])[["lifeExp","gdpPercap"]].agg(np.mean)
```

```python
groupby_reset = groupby.reset_index()
groupby_reset.head(10)
```
