### 1. Columns containing values, not variables ( Column headers are values, not variable names.)

```python
pew = pd.read_csv("../data/pew.csv")
pew.shape
```

```python
# pd.melt()
# id_vars - data taht stay the same
# values_vars - variables on which the data is going to melt on
# var_name - 
# value_name- by default columns using values name
```

```python
pew.melt(id_vars=["religion"])
```

```python
pew_tidy = pew.melt(id_vars=["religion"], var_name="income", value_name ="count") 
```

```python
pew_tidy.head()
```

```python
# now we can do some groupby 
import numpy as np
```

```python
# these all do the same thing
pew_tidy.groupby("income")["count"].agg("sum")
pew_tidy.groupby("income")["count"].sum()
pew_tidy.groupby("income")["count"].agg(np.sum)
```

```python
# Data set that has the same problem but more of it
```

```python
billboard = pd.read_csv("../data/billboard.csv")
billboard.head()
# for every week, how does the billboard ranking change over time?
```

```python
billboard_tidy = billboard.melt(id_vars=["year","artist", "track","time","date.entered"], var_name="week",value_name="rank")
billboard_tidy["week"]=billboard_tidy["week"].transform(trim)
billboard_tidy.sort_values(["artist","track"])
```

```python
def trim(name):
    return name[2:]
```

```python
billboard_tidy.groupby("artist")[["rank"]].mean()
```

### 3. Data is stored in the rows and in the columns

```python
weather = pd.read_csv("../data/weather.csv")
weather.head()
```

```python
weather_meal= weather.melt(id_vars =["id","year","month","element"],var_name="Day", value_name='temp' )
weather_meal
# the main indication that the data is stored in rows is if two rows share the same values up to that single one who creates that difference
```

```python
# pivot_tables
# index = columns that you don't toch
# columns _ which columns you want to pivot up
# values - values which have been used to populate rows of the columns that have been pivoted up

# by default pivot table has aggfunc and the reasone for this is, let say that some observation has been tocked from two man 
# and in this case we will get the main as the reasult.
```

```python
weather_tidy = weather_meal.pivot_table(index = ["id", "year", "month", "Day"], columns ="element", values = "temp").reset_index()
```

```python
weather_tidy.sort_values(["year", "month","Day"]).head()
```

```python
weather_tidy
```

```python
# if we don't care about the index we can rid of it through the process of saving the data
weather_tidy.to_csv("../data/weather_clean.csv", index = False)
```
