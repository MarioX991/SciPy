```python
ebola=pd.read_csv("../data/ebola_country_timeseries.csv")
ebola.head()
```

```python
ebola.columns
```

```python
ebola_melt=pd.melt(ebola,id_vars=["Date","Day"], var_name="cd_country",value_name="count")
ebola_melt
```

```python
var_split=ebola_melt["cd_country"].str.split("_")
var_split
```

```python
status_values=var_split.str.get(0)
cauntry_values=var_split.str.get(1)
```

```python
ebola_melt["status_vallues"]=status_values
ebola_melt["cauntry_values"]=cauntry_values
ebola_melt.head()
```

```python
# drugaciji nacin 

var_split=ebola_melt["cd_country"].str.split("_",expand=True)
var_split
```

```python
var_split.columns=["status1","status2"]
var_split.head()
```

```python
ebola_clean=pd.concat([ebola_melt,var_split],axis=1)
ebola_clean.drop(["cauntry_values","status_vallues"],axis=1,inplace=True)
```

```python
ebola_clean.head()
```

## Variables are stored in bot rows and columns

```python
weather=pd.read_csv("../data/weather.csv")
weather.head()
#dva ili vise redova su skoro po isti sem nekih malih promena
```

```python
weather_melt=pd.melt(weather,id_vars=["id","year","month","element"],
   var_name="day",value_name="temp")
weather_melt
```

```python
# hocemo svaki jedinstven element kolone da pretvorimo u kolonu
weather_tidy=weather_melt.pivot_table(
    index=["id","year","month","day"],
    columns="element",
    values="temp")
#index. - kolone koje jocemo da ostavimo iste
# columns - koju/e kolone hocemo da pivutujemo 
# values- cime hocemo da popunimo novonastalu kolunu
```

```python
weather_tidy.head()
```

```python
weather_flat=weather_tidy.reset_index()
weather_flat
```

```python
new_day=weather_flat['day'].str[1:].to_frame()
```
