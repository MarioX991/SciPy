## Koriscenje aggregate() ili agg()da bismo pozvali funkcije iz druge biblioteke

```python
# mozemo koristiti nase sopstvene funkcije u groupby 
def my_mean(values):
    n=len(values)
    s=np.sum(values)
    return s/n
```

```python
# ali ako vec funkcije postoje u nekoj od postojevcih biblioteka mozemo ih koristiti
gapminder.groupby("year")["lifeExp"].aggregate(np.mean)
```

```python
gapminder.groupby("year")["lifeExp"].agg(my_mean)
```

```python
# mozemo traziti da nam izbaci vise funkcija
gapminder.groupby("year")["lifeExp"].agg([np.count_nonzero, np.mean, np.std])
```

```python
gapminder.groupby(["year", "continent"])["lifeExp"].agg([np.count_nonzero, np.mean, np.std]).head()
```

```python
(gapminder.groupby(["year", "continent"])["lifeExp"]
 .agg([np.count_nonzero, np.mean, np.std])
 .rename(columns={"np.count_nonzero":"ncount",
                 "np.mean":"mean",
                 "np.std":"std"})).reset_index().head()
```
